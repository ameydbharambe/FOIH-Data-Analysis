```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.base.model import GenericLikelihoodModel
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming 'final_dataset1' is your DataFrame

# Drop rows with missing values in the selected columns
income_donations = final_dataset1[['HHINCOME', 'TOTGRNG']]
income_donations.dropna(subset=['HHINCOME', 'TOTGRNG'], inplace=True)
income_donations = income_donations[income_donations['HHINCOME'] > 0]

# Map ranges to their boundaries
range_to_bounds = {
    'NONE': (0, 0),
    '1-49 USD': (1, 49),
    '50-99 USD': (50, 99),
    '100-199 USD': (100, 199),
    '200-299 USD': (200, 299),
    '300-399 USD': (300, 399),
    '400-499 USD': (400, 499),
    '500-999 USD': (500, 999),
    '>1000 USD': (1000, np.inf)
}

income_donations['y_lower'], income_donations['y_upper'] = zip(*income_donations['TOTGRNG'].map(range_to_bounds))

# Define the likelihood function for interval regression
class IntervalRegression(GenericLikelihoodModel):
    def nloglikeobs(self, params):
        intercept, slope, sigma = params
        mu = intercept + slope * self.exog[:, 1]  # Accessing the HHINCOME values (exog[:, 1] because of add_constant)
        lower = (self.endog[:, 0] - mu) / sigma
        upper = (self.endog[:, 1] - mu) / sigma
        # Calculate the log-likelihood using scipy's norm
        log_likelihood = np.log(norm.cdf(upper) - norm.cdf(lower))
        return -log_likelihood

    def fit(self, start_params=None, **kwargs):
        if start_params is None:
            start_params = np.array([100, 0.0001, 1000])  # Adjusted initial guesses for intercept, slope, sigma
        return super().fit(start_params=start_params, **kwargs)

    def predict_donation(self, exog, params):
        intercept, slope, sigma = params
        exog = np.asarray(exog)  # Convert exog to a NumPy array
        predicted_donation = intercept + slope * exog[:, 1]  # Exog[:, 1] contains HHINCOME
        return predicted_donation

# Prepare the data for interval regression
exog = sm.add_constant(income_donations['HHINCOME'])
endog = income_donations[['y_lower', 'y_upper']].values

# Fit the interval regression model
interval_reg = IntervalRegression(endog=endog, exog=exog).fit()
print(interval_reg.summary())

# Predict donation amounts using the custom prediction method from the IntervalRegression class
income_donations['Predicted Donation'] = interval_reg.model.predict_donation(exog, interval_reg.params)

# Calculate the observed midpoints of donation ranges
income_donations['Observed Midpoint'] = (income_donations['y_lower'] + income_donations['y_upper']) / 2

# Visualization: Observed vs. Predicted Donation Ranges with Error Bars
plt.figure(figsize=(10, 6))
plt.errorbar(income_donations['HHINCOME'], income_donations['Observed Midpoint'],
             yerr=[income_donations['Observed Midpoint'] - income_donations['y_lower'],
                   income_donations['y_upper'] - income_donations['Observed Midpoint']],
             fmt='o', label='Observed Donation Ranges', alpha=0.6)

plt.scatter(income_donations['HHINCOME'], income_donations['Predicted Donation'],
            color='red', label='Predicted Donation', alpha=0.6)
plt.plot(np.unique(income_donations['HHINCOME']),
         np.poly1d(np.polyfit(income_donations['HHINCOME'], income_donations['Predicted Donation'], 1))
         (np.unique(income_donations['HHINCOME'])),
         color='red', linestyle='--', label='Trend Line (Predicted)')

plt.title('Observed vs. Predicted Donation Ranges by Income Level')
plt.xlabel('Income')
plt.ylabel('Donation Amount')
plt.legend()
plt.grid(True)
plt.show()
```
